```python
class KVPair:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        
    def __repr__(self):
        return f'[key: {self.key}, val: {self.value}]'
```

```python
import os
import json

        
class DayByDayData(list):
    def __init__(self, source_file, source_dir='data', treshold=0):
        super().__init__()
        with open(f'{source_dir}/{source_file}', 'r') as sf:
            json_data = json.load(sf)
        for key in json_data:
            if json_data[key] > treshold:
                self.append(KVPair(key, json_data[key]))
        
        self.sort(reverse=True, key=lambda x: x.value)
        

        
files = os.listdir('data')
obj_list = [DayByDayData(file, treshold=1) for file in files]
print(obj_list[0][0:10])
```

```python
class DatasetHandler(list):
    def __init__(self, dataset):
        super().__init__()
        self.id_history = {card_id: [] for card_id in [x.key for x in dataset[0]]}
        for record in dataset:
            self.append([x.value for x in record])
            for number, item in enumerate(record):
                try:
                    self.id_history[item.key].append(number)
                except KeyError:
                    self.id_history[item.key] = [number]
        
    def get_nth_card(self, n):
        return [record[n] for record in self]
    
    def save(self, output_file, dump_dict=None):
        if dump_dict is None:
            dump_dict = {'main': [x for x in self], 'price_history': self.id_history}
            
        with open(output_file, 'w') as of:
            json.dump(dump_dict, of)
            
    
    
dh = DatasetHandler(obj_list)


import matplotlib.pyplot as plt 

for i in range(50, 100, 5):
    plt.plot(dh.get_nth_card(i))

plt.show()
dh.save('test_dump.json')
```
